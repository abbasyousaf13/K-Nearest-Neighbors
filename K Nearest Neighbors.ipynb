# 2nd part

### divide your data into two parts train and test for cross validation

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

# 3rd step

### Standard scaling
### because the difference between age and estimated salary column is very huge

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_train
```

```python
X_test=scaler.transform(X_test)
X_test
```

# Applying KNN

### there are two methods to calculate the value of K
### 1st method - simply find out the square root

```python
np.sqrt(X_train.shape[0])      # there are 320 rows in first index
```

```python
# we will assume the value of K is equals to 17
k=17
```

```python
# import KNN class
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=k)
```

```python
# train our model
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
y_pred.shape
```

```python
y_test.shape
```

```python
# Now we just compare out y_pred to our y_test to check our accuracy and we will see how accurate our model was
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# we can also calculate the confusion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,y_pred)
```

### 2nd method - run the loop and create the object everytime

```python
accuracy=[]

for i in range(1,26):
    knn=KNeighborsClassifier(n_neighbors=i) # current iteration of the loop
    knn.fit(X_train,y_train)      # everytime fit X_train and y_train / training the algorithm everytime the loop is running
    accuracy.append(accuracy_score(y_test,knn.predict(X_test)))
```

```python
accuracy
```

```python
len(accuracy)
```

```python
# plotting the graph to see where we are achieving the highest accuracy
plt.plot(range(1,26),accuracy)
```

```python
def predict_output():
    age=int(input("Enter the age "))
    salary=int(input("Enter the salary "))
    
    # return=np.array([[age],[salary]]).shape
    X_new=np.array([[age],[salary]]).reshape(1,2)  # we are reshaping it because our original X_train and X_test had 2 columns 
    # but over here we have 1 column. So whenerver we are providing input to our classifier it should have 2 columns
    X_new=scaler.transform(X_new)
    
    if knn.predict(X_new)[0]==0:
        return "Will purchase"
    else:
        return "Will not purchase"
```
